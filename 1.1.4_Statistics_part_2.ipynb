````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''
    
    N = len(X)
    sum_of_x = sum(X)
    sum_of_y = sum(Y)
    sum_of_xy = sum([xi*yi for xi, yi in zip(X, Y)])
    sum_of_x_sq_2 = sum(xi**2 for xi in X)
    
    numer = (N*sum_of_xy) - (sum_of_x * sum_of_y)
    denum = (N*sum_of_x_sq_2) - (sum_of_x ** 2)
    
    m = numer / denum
    b = (sum_of_y - (m*sum_of_x))/N
    print(m, ' and ', b)
    print('y = {:.2f}x + {:.2f}' . format(m,b))
    return m, b

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
# ANSWER
plt.scatter(X,Y)
yfit = [b + m * xi for xi in X]
plt.plot(X, yfit)
plt.grid()
plt.show()
````
